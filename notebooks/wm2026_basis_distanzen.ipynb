# WM 2026 – Der unsichtbare Gegner
## Erstellt Basis-Datenstruktur der 16 WM-Stadien + berechnet Distanzen zwischen Stadien


```python
import pandas as pd
from math import radians, sin, cos, sqrt, asin
```

Für die Berechnung der Distanzen wird die  Haversine-Formel genutzt (Luftlinie)




```python
from pathlib import Path

def haversine_distance(lat1, lon1, lat2, lon2):
    """Berechnet Luftlinien-Distanz in km zwischen zwei GPS-Koordinaten"""
    R = 6371  # Erdradius in km
    lat1, lon1, lat2, lon2 = radians(lat1), radians(lon1), radians(lat2), radians(lon2)
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    return R * c

# 16 WM-Stadien mit Koordinaten
stadien_data = {
    "Stadium_Name": [
        "AT&T Stadium", "BC Place", "BMO Field", "Estadio Akron", "Estadio Banorte",
        "Estadio BBVA", "GEHA Field at Arrowhead Stadium", "Gillette Stadium",
        "Hard Rock Stadium", "Levi's Stadium", "Lincoln Financial Field", "Lumen Field",
        "Mercedes-Benz Stadium", "MetLife Stadium", "NRG Stadium", "SoFi Stadium"
    ],
    "City": [
        "Dallas/Arlington", "Vancouver", "Toronto", "Guadalajara", "Mexico City",
        "Monterrey", "Kansas City", "Boston", "Miami", "San Francisco",
        "Philadelphia", "Seattle", "Atlanta", "New York/New Jersey", "Houston", "Los Angeles"
    ],
    "Country": [
        "USA", "Canada", "Canada", "Mexico", "Mexico", "Mexico",
        "USA", "USA", "USA", "USA", "USA", "USA", "USA", "USA", "USA", "USA"
    ],
    "Latitude": [
        32.748138, 49.276646, 43.633087, 20.681721, 19.302837,
        25.669132, 39.048855, 42.090790, 25.957830, 37.403297,
        39.901325, 47.595135, 33.755371, 40.813477, 29.684702, 33.953438
    ],
    "Longitude": [
        -97.093231, -123.112564, -79.418961, -103.463135, -99.150803,
        -100.244621, -94.484474, -71.264404, -80.239326, -121.969765,
        -75.167862, -122.331917, -84.401436, -74.074951, -95.410965, -118.339447
    ],
    "Timezone_UTC": [
        -5, -4, -4, -6, -6, -6, -5, -4, -5, -7, -5, -7, -5, -4, -6, -7
    ],
    "Matches_Count": [
        9, 7, 6, 4, 5, 4, 6, 7, 7, 6, 6, 6, 8, 8, 7, 8
    ]
}

df_stadien = pd.DataFrame(stadien_data)

# CSV in den data/-Ordner speichern (absoluter Pfad, damit es egal ist,
# welches Arbeitsverzeichnis der Kernel hat)
DATA_DIR = Path("/Users/christine/Documents/03 Datenprojekte/WM2026_Analyse/data")
DATA_DIR.mkdir(parents=True, exist_ok=True)

csv_pfad = DATA_DIR / "wm2026_stadien_koordinaten_zeitzonen.csv"
df_stadien.to_csv(csv_pfad, index=False, encoding="utf-8-sig")

print("=== Basis-Datenstruktur erstellt ===")
print(f"{len(df_stadien)} Stadien in {df_stadien['Country'].nunique()} Ländern")
print(f"CSV gespeichert: {csv_pfad}")

# Beispiel-Distanzen
print("\n=== Beispiel-Distanzen ===")
distanzen = [
    ("Dallas", "Mexico City", 0, 4, "1.300 km"),
    ("Vancouver", "LA", 1, 15, "3.400 km"),
    ("Boston", "Atlanta", 7, 12, "1.500 km")
]
for c1, c2, i1, i2, _ in distanzen:
    d = haversine_distance(
        df_stadien.loc[i1, 'Latitude'], df_stadien.loc[i1, 'Longitude'],
        df_stadien.loc[i2, 'Latitude'], df_stadien.loc[i2, 'Longitude']
    )
    print(f"{c1} → {c2}: {d:.0f} km")
```
